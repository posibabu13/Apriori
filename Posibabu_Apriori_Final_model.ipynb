```python
Apriori_Australia=Apriori[Apriori['Country'] =="Australia"] .groupby(['InvoiceNo', 'Description'])['Quantity'] .sum().unstack(1).fillna(0)
```

```python
Apriori_Australia
```

```python
Apriori_Australia.reset_index()
```

```python
def record_value(x):
    if(x<=0):
        return 0
    if(x>=1):
        return 1
```

```python
Apriori_Australia=Apriori_Australia.applymap(record_value)
```

```python
Apriori_Australia.head(4)
```

### Building the Model

```python
# Building the model
frq_items = apriori(Apriori_Australia, min_support = 0.07, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```
